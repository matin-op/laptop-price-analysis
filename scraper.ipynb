```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from sqlalchemy import create_engine
```

```python
url= "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
products = soup.find_all("div", class_="col-md-4 col-xl-4 col-lg-4")
data=[]
# print(len(products))
for product in products:
    # title= product.select_one("h4",class_=["title","card-title"]).text.strip()
    title = product.select_one("a.title").text.strip()
    price = product.select_one("h4.price.float-end.card-title").text.strip()
    rating= len(product.find(class_="ratings").find_all("span", class_=["ws-icon", "ws-icon-star"]))
    review= product.find(class_="review-count").text.strip()
										
    data.append({
        "title": title,
        "price":price,
        "rating":rating,
        "reviews":review})
df= pd.DataFrame(data)


```
